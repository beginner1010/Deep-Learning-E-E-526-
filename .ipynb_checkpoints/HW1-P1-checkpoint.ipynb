```python
from __future__ import division, print_function, unicode_literals
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
rc('text', usetex = True)
rc('font', family='serif')
```

### Get the data:

```python
df = pd.read_csv('cars.csv', sep=',')
df.rename(index=str, columns={'Unnamed: 0' : 'ID'}, inplace=True)
df.head()
```

### The goal is to predict distance. So $X$ is speed of the car and $y$ is distance.

```python
X, y = df.speed, df.dist
```

```python
def predict_y(X, coeff):
    return X.dot(coeff)
```

### Change $X$ and add polynomial feature based on the given power $n$:

```python
def transform(X, n):
    data_X = np.c_[X]
    X = np.c_[np.ones((len(X), 1)), X]
    for i in range (2, n + 1):
        X = np.c_[X, data_X ** i]
    return X
```

### Normal Equation:

```python
def linear_regression(X, y):
    opt_coeffs = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return opt_coeffs
```

# Plot
### $\theta$ for different polynomial features are printed below.
### The result shows that degree $4$ causes overfitting.

```python
ax = plt.figure(figsize=(8, 6))
plt.plot(X, y, 'k.', label='_nolegend_', markersize=12, alpha=0.6)
for poly in range (1, 5):
    X_poly = transform(X, poly)
    coeff = linear_regression(X_poly, y)
    
    X_test = np.linspace(0, 30, 200)
    X_test_poly = transform(X_test, poly)
    y_pred = predict_y(X_test_poly, coeff)
    
    plt.plot(X_test, y_pred, '-',label=("Degree %d" % poly), 
             linewidth=4, alpha=0.6)
    equation = 'y = '
    for i in range (0, poly + 1):
        if i != 0: equation += ' + '
        equation += ('%.3f x^%d' % (coeff[i], i))
    print('For degree %d we have:' % (poly), equation)
plt.legend(fontsize=18, fancybox=True, framealpha=1, 
           shadow=True, borderpad=1)
plt.xlabel('Speed')
dump = plt.ylabel('Distance')
```
